```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("Datasets/WineQualityReds.csv")
```

```python
df.drop(['Unnamed: 0'], axis='columns', inplace=True)
```

```python
df.rename(columns = {'fixed.acidity': 'facidity', 'volatile.acidity': 'vacidity',
                     'citric.acid':'cacid','total.sulfur.dioxide':'tsdioxide',
                     'residual.sugar':'rsugar','free.sulfur.dioxide':'sdioxide'}, inplace = True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[["facidity"]] = scaler.fit_transform(df[["facidity"]])
df[["vacidity"]] = scaler.fit_transform(df[["vacidity"]])
df[["cacid"]] = scaler.fit_transform(df[["cacid"]])
df[["rsugar"]] = scaler.fit_transform(df[["rsugar"]])
df[["chlorides"]] = scaler.fit_transform(df[["chlorides"]])
df[["sdioxide"]] = scaler.fit_transform(df[["sdioxide"]])
df[["tsdioxide"]] = scaler.fit_transform(df[["tsdioxide"]])
df[["pH"]] = scaler.fit_transform(df[["pH"]])
df[["sulphates"]] = scaler.fit_transform(df[["sulphates"]])
df[["alcohol"]] = scaler.fit_transform(df[["alcohol"]])
df[["quality"]] = scaler.fit_transform(df[["quality"]])
```

```python
X = df.drop('quality', axis='columns')
X.head()
```

```python
Y =df['quality']
Y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```
